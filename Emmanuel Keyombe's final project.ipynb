```python
#importing neccessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('Customers.csv',)
```

```python
data
```

```python
data.info()
```

```python
#stats summarry
data.describe()
```

```python
#summary stats
import numpy as np
mean = np.mean(data['Age'])
median = np.median(data['Age'])
std_deviation = np.std(data['Age'])
#print(mean)
print(f"Mean : {mean}")
print(f"Median : {median}")
print(f"Standard Deviation : {std_deviation}")
```

```python
#
plt.hist(data['Age'],bins=10,color='blue',edgecolor='black')
plt.title('Age')
plt.ylabel('frequency')
plt.xlabel('number of people')
plt.show()
```

```python
plt.hist(data['Age'],bins=10,color='blue',edgecolor='black')
plt.title('Age')
plt.ylabel('frequency')
plt.xlabel('number of people')

plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame(data)
```

```python
#Relationship between Age and Work Experience
correlation = df['Age'].corr(df['Work Experience'])
correlation
```

```python
# Visual Relationship
sns.scatterplot(data=df, x='Age', y='Work Experience')

plt.xlabel('Age')
plt.ylabel('Work Experience')
plt.title('Relationship Between Age and Work Experience')
plt.tight_layout()
plt.show()
```
